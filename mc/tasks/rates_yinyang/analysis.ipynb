```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
results = np.load("./results_data/results_run_time.npz")
```

```python
results.files
```

```python
acc = results["acc"]
loss = results["loss"]
run_time = results["run_time"]

epoch_ax = results["epoch_ax"]
```

```python
epoch_ax
```

```python
acc_mean = acc.mean(axis=1)
acc_std = acc.std(axis=1)

plt.fill_between(x=epoch_ax, y1=np.arange(acc_mean.shape[0]), y2=acc_mean-acc_std, acc_mean+acc_std, alpha=0.25)
plt.plot(acc_mean)

plt.show()
```

```python
loss_mean = loss.mean(axis=1)
loss_std = loss.std(axis=1)

plt.fill_between(np.arange(loss_mean.shape[0]), loss_mean - loss_std, loss_mean + loss_std, alpha=0.25)
plt.plot(loss_mean)

plt.show()
```
